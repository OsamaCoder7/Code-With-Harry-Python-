## Enumerate Function

```python
marks = [12,54,56,77,85,24,35]
for index, mark in enumerate(marks):
    print(f"{index}: {mark}")
    if (index == 3):
        print("Awesome!!")
```

```python
players = ["Imam", "Fakhar", "Babar", "Rizwan", "Salman", "Iftikhar", "Shadab","Faheem", "Shaheen", "Rauf", "Abrar"]
for index, player in enumerate(players):
    print(f"{index}:{player}")
```

## Global and Local Variables

```python
x = 4
print(x)

def hello():
    x = 5
    print(f"The local x is: {x}")
    print("Hello Emaz")

print(f"The Global x is: {x}")    
print("fg")
hello()    
x = 5
print(f"The Global x is: {x}")
```

```python
x = 10
def myfunc():
    global x
    x = 9
    y = 7
    print(y)
myfunc()
print(x)
```

```python
x = 18
def myfunc():
    global x
    x = 333
    y = 56
    print(f"Babar's Shirt num is {y}")
myfunc()
print(f"Gayle's Shirt num is {x}")
```

## File Io in Python

```python
f = open("main.txt", "rb")               ## binary form
print(f)
text = f.read()
print(text)
f.close()
```

```python
f = open("main.txt", "r")               ## read form
print(f)
text = f.read()
print(text)
f.close()
```

```python
f = open("main.txt", "a")               ## write form
f.write("This is the new content of the file\n")
f.close()


with open("main.txt", "a") as f:
    f.write("Hey I am inside with Emaz")
    f.close()    
```
